```python
import pandas as pd
```

```python
math_df = pd.read_csv('../../data/processed/math.csv', delimiter = ';', encoding = 'utf-8')
inf_df = pd.read_csv('../../data/processed/informatics.csv', delimiter = ';', encoding = 'utf-8')
```

### Exploratory analysis

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
math_df['shortcut'].value_counts().plot(kind='bar')
plt.title('Math - Distribution of examples across shortcut')
plt.xlabel('Shortcut')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
inf_df['shortcut'].value_counts().plot(kind='bar')
plt.title('Informatics - Distribution of examples across shortcut')
plt.xlabel('Shortcut')
plt.ylabel('Count')
plt.show()
```

```python
import json

unique_texts = set()
for q in inf_df['question'].dropna():
    try:
        parsed = json.loads(q)
        for item in parsed:
            if isinstance(item, list) and len(item) > 1 and item[0] == 'text':
                unique_texts.add(item[1])
    except:
        pass

print("Unique text values from question column:")
for t in unique_texts:
    print(t)
```
